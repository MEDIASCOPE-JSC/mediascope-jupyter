## Начало работы

Положите в одну директорию JupiterHub все файлы из архива `Share_of_Search.zip`:
- `FileApi_view.ipynb` – блокнот для просмотра доступных файлов с помощью FileApi;
- `FileApi_download.ipynb`  – блокнот для скачивания архива с csv-файлами с помощью FileApi;
- `FileApi_calculation_categories.ipynb` – блокнот с примерами расчета показателей в разбивке по категориям;
- `FileApi_calculation_brands.ipynb` – блокнот с примерами расчета показателей в разбивке по брендам;
- `FileApi_calculation_brands_SoS.ipynb` – блокнот с примером расчета показателя SoS;
- `FileApi_calculation_demography.ipynb` – блокнот с примером расчета социально-демографической структуры аудитории по брендам;
- `FileApi_calculation_demography_SoS.ipynb` – блокнот с примером расчета показателя SoS по соц.-дем параметрам;
- `help.ipynb` – текущий блокнот со справочными материалами;
- `keys_file_api.txt` – блокнот с учетными данными для получения токена;
- `fileapiback.py` – вспомогательный блокнот с функциями для работы блокнотов. Его открывать и использовать не нужно.

## Аутентификация и авторизация в API

### Логин и пароль

Для работы с FileApi вам были выданы: логин, пароль и название агентства. Эти данные требуются для получения API-токена.\
Получение токена осуществляется автоматически при работе с блокнотами (в блокнотах с примерами расчета показателей токен не требуется).

Перед началом работы с API добавьте персональные данные в файл `keys_file_api.txt`.\
Откройте файл `keys_file_api.txt` и укажите название агентства, логин и пароль:

	agency:agency
    username:username
	password:password

Сохраните файл.

**Важно**: 
- вводите название агенства, логин, пароль именно в том виде, в котором вы их получили от клиентского менеджера Mediascope;
- не добавляйте лишние пробелы и знаки. Например, пароль должен быть записан так: password=12345.
