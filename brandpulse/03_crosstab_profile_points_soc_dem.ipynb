```python
profile_id = 3900499002973208128 # id профиля 2024

# задаем параметры целевых групп (используя sys_name из базы)
demo = [
    {
        "demo_name": "M12_24",
        "demo_param": [
            {"property_name": "SEX", 
            "expression": "='MALE'"},
            {"property_name": "AGE_COUNT2", 
            "expression": " BETWEEN 'A12' AND 'A24' "}
        ]
    },
    {
        "demo_name": "M25_44",
        "demo_param": [
            {"property_name": "SEX", 
            "expression": "='MALE'"},
            {"property_name": "AGE_COUNT2", 
            "expression": " BETWEEN 'A25' AND 'A44' "}
        ]
    },
    {
        "demo_name": "M45_64",
        "demo_param": [
            {"property_name": "SEX", 
            "expression": "='MALE'"},
            {"property_name": "AGE_COUNT2", 
            "expression": " BETWEEN 'A45' AND 'A64' "}
        ]
    },
    {
        "demo_name": "Ж12_24",
        "demo_param": [
            {"property_name": "SEX", 
            "expression": "='FEMALE'"},
            {"property_name": "AGE_COUNT2", 
            "expression": " BETWEEN 'A12' AND 'A24' "}
        ]
    },
    {
        "demo_name": "Ж25_44",
        "demo_param": [
            {"property_name": "SEX", 
            "expression": "='FEMALE'"},
            {"property_name": "AGE_COUNT2", 
            "expression": " BETWEEN 'A25' AND 'A44' "}
        ]
    },
    {
        "demo_name": "Ж45_64",
        "demo_param": [
            {"property_name": "SEX", 
            "expression": "='FEMALE'"},
            {"property_name": "AGE_COUNT2", 
            "expression": " BETWEEN 'A45' AND 'A64' "}
        ]
    }
]
```

```python
# функции для облегчения построения датафреймов для всех демогрупп
from typing import Callable, Optional
import time
import sys
import os
import logging
from concurrent.futures import ThreadPoolExecutor, as_completed
from threading import Lock
import multiprocessing

# Настройка логирования для Jupyter
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s',
    handlers=[logging.StreamHandler(sys.stdout)]
)
logger = logging.getLogger(__name__)

def process_single_demogroup(
    idx: int,
    total_demo: int,
    demo_item: dict,
    project_id: int,
    properties: list,
    universe_func: Callable,
    logger_lock: Lock
) -> tuple:
    """
    Обработка одной демографической группы (для использования в потоках)
    
    Returns:
        tuple: (df_tmp, demo_stats_dict)
    """
    demo_start_time = time.time()
    demo_name = demo_item['demo_name']
    
    # Логируем начало обработки с блокировкой
    with logger_lock:
        logger.info(f"[{idx}/{total_demo}] Обработка демогруппы: {demo_name}")
    
    try:
        # Получение данных
        df_tmp = (
            universe_func(
                project_ids=[project_id], 
                demo=demo_item['demo_param'], 
                properties=properties,
                #log_sql=True
            )
            .reset_index(drop=True)
            .rename(columns={
                "universe": f"{demo_name}_universe", 
                "sample": f"{demo_name}_sample"
            })
        )
        
        # Статистика по времени
        demo_time = time.time() - demo_start_time
        
        demo_stats = {
            'name': demo_name,
            'time': demo_time,
            'rows': len(df_tmp),
            'cols': len(df_tmp.columns)
        }
        
        # Логирование результатов с блокировкой
        with logger_lock:
            logger.info(f"  ✓ [{idx}/{total_demo}] {demo_name} завершено за {demo_time:.2f} сек")
        
        return df_tmp, demo_stats
        
    except Exception as e:
        with logger_lock:
            logger.error(f"  ✗ Ошибка при обработке демогруппы {demo_name}: {str(e)}", exc_info=True)
        raise

# ==================== ОСНОВНАЯ ФУНКЦИЯ ====================

def calc_project_universe(
    project_id: int,
    properties: list,
    demo: list[dict],
    universe_func: Callable,
    total_func: Callable,
    log_level: str = "INFO",
    max_workers: Optional[int] = MAX_WORKERS,
    use_threading: bool = True
) -> pd.DataFrame:
    """
    Универсальная функция для расчета project universe (property или category).
    
    Args:
        project_id: ID проекта
        properties: список свойств
        demo: список демографических групп
        universe_func: функция для расчета universe по демогруппам
        total_func: функция для расчета total
        log_level: уровень логирования ("DEBUG", "INFO", "WARNING", "OFF")
        max_workers: максимальное количество потоков (None = auto)
        use_threading: использовать многопоточность (True) или последовательную обработку (False)
    
    Returns:
        pd.DataFrame: объединенный датафрейм
    """
    # Устанавливаем уровень логирования
    if log_level == "DEBUG":
        logger.setLevel(logging.DEBUG)
    elif log_level == "INFO":
        logger.setLevel(logging.INFO)
    elif log_level == "WARNING":
        logger.setLevel(logging.WARNING)
    elif log_level == "OFF":
        logger.setLevel(logging.CRITICAL + 10)
    
    func_name = universe_func.__name__
    start_time = time.time()
    
    # Определяем количество потоков
    if max_workers is None:
        # По умолчанию используем количество CPU ядер, но не больше количества демогрупп
        max_workers = min(multiprocessing.cpu_count(), len(demo))
        # Ограничиваем сверху для избежания слишком большого количества потоков
        max_workers = min(max_workers, MAX_WORKERS)
        
    logger.info(f"Начало расчета {properties}")
    
    data = []
    demo_stats = []
    logger_lock = Lock()
    
    if use_threading and len(demo) > 1:
        # Многопоточная обработка демогрупп
        
        with ThreadPoolExecutor(max_workers=max_workers) as executor:
            # Создаем задачи для всех демогрупп
            futures = []
            for idx, d in enumerate(demo, 1):
                future = executor.submit(
                    process_single_demogroup,
                    idx, len(demo), d, project_id, properties, 
                    universe_func, logger_lock
                )
                futures.append((idx, future))
            
            # Собираем результаты по мере завершения
            completed = 0
            for idx, future in futures:
                try:
                    df_tmp, demo_stat = future.result(timeout=10800)  # Таймаут 3 часа на демогруппу
                    data.append(df_tmp)
                    demo_stats.append(demo_stat)
                    completed += 1
                            
                except Exception as e:
                    with logger_lock:
                        logger.error(f"Ошибка в потоке для демогруппы #{idx}: {str(e)}", exc_info=True)
                    raise
    else:
        # Последовательная обработка демогрупп (для обратной совместимости)
        logger.info("Запуск последовательной обработки")
        
        for idx, d in enumerate(demo, 1):
            df_tmp, demo_stat = process_single_demogroup(
                idx, len(demo), d, project_id, properties, 
                universe_func, logger_lock
            )
            data.append(df_tmp)
            demo_stats.append(demo_stat)
    
    if not data:
        logger.warning(f"Нет данных для {func_name}")
        return pd.DataFrame()
    
    # Объединение данных по демогруппам
    key_cols = data[0].columns[:-2].tolist()
    logger.debug(f"Ключевые колонки для объединения: {key_cols}")
    
    universe = reduce(lambda left, right: pd.merge(left, right, on=key_cols, how='outer'), data)
    
    try:
        total = (total_func(
            project_ids=[project_id], 
            properties=properties, 
            demo=None
        )
        .reset_index(drop=True)
        .rename(columns={"universe": "total universe"}))
        
    except Exception as e:
        logger.error(f"Ошибка при расчете total: {str(e)}", exc_info=True)
        raise
    
    # Приведение датафреймов к единому виду
    logger.debug("Приведение структуры данных к единому виду")
    
    if 'category_sys_name' not in total.columns:
        logger.debug("Добавление колонок category_sys_name и category_text")
        universe.insert(
            loc=universe.columns.get_loc('property_text') + 1, 
            column='category_sys_name', 
            value=''
        )
        universe.insert(
            loc=universe.columns.get_loc('category_sys_name') + 1, 
            column='category_text', 
            value=''
        )
        
        total.insert(
            loc=total.columns.get_loc('property_text') + 1, 
            column='category_sys_name', 
            value=''
        )
        total.insert(
            loc=total.columns.get_loc('category_sys_name') + 1, 
            column='category_text', 
            value=''
        )
    
    if 'brand_sys_name' not in total.columns:
        logger.debug("Добавление колонок brand_sys_name и brand_text")
        universe.insert(
            loc=universe.columns.get_loc('category_text') + 1, 
            column='brand_sys_name', 
            value=''
        )
        universe.insert(
            loc=universe.columns.get_loc('brand_sys_name') + 1, 
            column='brand_text', 
            value=''
        )
        
        total.insert(
            loc=total.columns.get_loc('category_text') + 1, 
            column='brand_sys_name', 
            value=''
        )
        total.insert(
            loc=total.columns.get_loc('brand_sys_name') + 1, 
            column='brand_text', 
            value=''
        )
    
    merged = universe.merge(
        total, 
        on=['property_sys_name', 'property_text', 'category_sys_name', 'brand_sys_name'], 
        how='left'
    )
    
    merged.rename(columns={
        "category_text_x": "category_text", 
        "brand_text_x": "brand_text"
    }, inplace=True)
    
    # Формирование результата
    key_cols = merged.columns.to_list()[:8]
    demo_cols = universe.columns.to_list()[8:]
    result = merged[key_cols + ['total universe', 'sample'] + demo_cols].copy()
    
    # Итоговая статистика
    total_time = time.time() - start_time
    logger.info(f"Завершение расчета {properties}")
    logger.info(f"Общее время выполнения: {total_time:.2f} сек")
    
    return result

# ==================== ОБЕРТКИ ДЛЯ КОНКРЕТНЫХ ТИПОВ ====================

def calc_category(properties: list[tuple], option_text: str = None, 
                  log_level: str = "INFO", max_workers: Optional[int] = None, use_threading: bool = True):
    """Расчет для категорий"""
    
    try:
        result = calc_project_universe(
            properties=properties, 
            project_id=profile_id, 
            demo=demo, 
            universe_func=bp.calc_project_category_universe,
            total_func=bp.calc_universe_property_by_category,
            log_level=log_level,
            max_workers=max_workers,
            use_threading=use_threading
        )
        
        return result
        
    except Exception as e:
        logger.error(f"Ошибка в calc_category: {str(e)}", exc_info=True)
        raise

def calc_brand(properties: tuple, option_text: str = None, 
               log_level: str = "INFO", max_workers: Optional[int] = None, use_threading: bool = True):
    """Расчет для брендов"""
    
    try:
        result = calc_project_universe(
            properties=properties, 
            project_id=profile_id, 
            demo=demo, 
            universe_func=bp.calc_project_brand_universe,
            total_func=bp.calc_universe_property_by_brand,
            log_level=log_level,
            max_workers=max_workers,
            use_threading=use_threading
        )
        
        return result
        
    except Exception as e:
        logger.error(f"Ошибка в calc_brand: {str(e)}", exc_info=True)
        raise

def calc_property(properties: tuple, option_text: str = None, 
                  log_level: str = "INFO", max_workers: Optional[int] = None, use_threading: bool = True):
    """Расчет для свойств"""    
    try:
        result = calc_project_universe(
            properties=properties, 
            project_id=profile_id, 
            demo=demo, 
            universe_func=bp.calc_project_property_universe,
            total_func=bp.calc_universe_property,
            log_level=log_level,
            max_workers=max_workers,
            use_threading=use_threading
        )
        
        return result
        
    except Exception as e:
        logger.error(f"Ошибка в calc_property: {str(e)}", exc_info=True)
        raise
```

```python
# добавляем строку сводных итогов
full = pd.concat([df_dosug, df_zdor, df_inet, df_adv, df_retail, df_trans], ignore_index=True)

non_num_cols = full.columns[:8]

total_row = {
    non_num_cols[0]: 'total',
    non_num_cols[1]: '',
    non_num_cols[2]: '',
    non_num_cols[3]: '',
    non_num_cols[4]: '',
    non_num_cols[5]: '',
    non_num_cols[6]: '',
    non_num_cols[7]: '',
}

total_row['total universe'] = bp.calc_universe(project_ids=[profile_id], demo=None)
total_row['sample'] = bp.calc_sample(project_ids=[profile_id], demo=None)
total_row['col%'] = 100

for d in demo:
    total_row[d['demo_name']+'_universe'] = bp.calc_universe(project_ids=[profile_id], demo=d['demo_param'])
    total_row[d['demo_name']+'_sample'] = bp.calc_sample(project_ids=[profile_id], demo=d['demo_param'])
    total_row[d['demo_name']+'_col%'] = 100
    total_row[d['demo_name']+'_row%'] = 100 * total_row[d['demo_name']+'_universe'] / total_row['total universe']
    total_row[d['demo_name']+'_affinity'] = 100
total_row
```

Добавляем к рассчитанному universe расчет долей (row% и col%) и аффинити

```python
# добавляем доли и аффинити
# считаем доли
result_tmp = full[non_num_cols].copy()

result_tmp['total universe'] = full['total universe']
#result_tmp['sample'] = full['sample']
result_tmp['col%'] = 100 * result_tmp['total universe'] / total_row['total universe']

for d in demo:
    result_tmp[d['demo_name']+'_universe'] = full[d['demo_name']+'_universe']
    result_tmp[d['demo_name']+'_sample'] = full[d['demo_name']+'_sample']
    result_tmp[d['demo_name']+'_col%'] = 100 * full[d['demo_name']+'_universe'] / total_row[d['demo_name']+'_universe']
    result_tmp[d['demo_name']+'_row%'] = 100 * full[d['demo_name']+'_universe'] / result_tmp['total universe']
    result_tmp[d['demo_name']+'_affinity'] = 100 * result_tmp[d['demo_name']+'_col%'] / result_tmp['col%']

total_row_df = pd.DataFrame([total_row], columns=result_tmp.columns)
result = pd.concat([total_row_df, result_tmp], ignore_index=True)

# удаляем лишние колонки
result.drop(columns=[d['demo_name'] + '_sample' for d in demo], inplace=True)

# округляем до нужной точности
result['col%'] = result['col%'].round(1)
for d in demo:
    result[d['demo_name']+'_col%'] = result[d['demo_name']+'_col%'].round(1)
    result[d['demo_name']+'_row%'] = result[d['demo_name']+'_row%'].round(1)
    result[d['demo_name']+'_affinity'] = result[d['demo_name']+'_affinity'].round(0)

result
```
