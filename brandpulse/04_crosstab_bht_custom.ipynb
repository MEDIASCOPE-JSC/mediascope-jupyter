Рассчитаем сводные тоталы

```python
# добавляем строку сводных итогов
non_num_cols = df_result.columns[:8]

total_universe = bp.calc_universe(project_ids=[project_id], demo=None)
total_sample = bp.calc_sample(project_ids=[project_id], demo=None)

total_row = {
    non_num_cols[0]: 'total',
    non_num_cols[1]: '',
    non_num_cols[2]: '',
    non_num_cols[3]: '',
    non_num_cols[4]: '',
    non_num_cols[5]: '',
    non_num_cols[6]: total_universe,
    non_num_cols[7]: total_sample,
    'col%': 100    
}

for d in demo:
    total_row[d['demo_name']+'_universe'] = bp.calc_universe(project_ids=[project_id], demo=d['demo_param'])
    total_row[d['demo_name']+'_sample'] = bp.calc_sample(project_ids=[project_id], demo=d['demo_param'])
    total_row[d['demo_name']+'_col%'] = 100
    total_row[d['demo_name']+'_row%'] = round(100 * total_row[d['demo_name']+'_universe'] / total_row[non_num_cols[6]], 1)
    total_row[d['demo_name']+'_affinity'] = 100
total_row
```

Добавляем к рассчитанному universe расчет долей (row% и col%) и аффинити

```python
# считаем доли и аффинити    
result_tmp = df_result[non_num_cols].copy()

result_tmp['col%'] = 100 * df_result['total universe'] / total_row['total universe']

for d in demo:
    result_tmp[d['demo_name']+'_universe'] = df_result[d['demo_name']+'_universe']
    result_tmp[d['demo_name']+'_sample'] = df_result[d['demo_name']+'_sample']
    result_tmp[d['demo_name']+'_col%'] = 100 * df_result[d['demo_name']+'_universe'] / total_row[d['demo_name']+'_universe']
    result_tmp[d['demo_name']+'_row%'] = 100 * df_result[d['demo_name']+'_universe'] / df_result['total universe']
    result_tmp[d['demo_name']+'_affinity'] = 100 * result_tmp[d['demo_name']+'_col%'] / result_tmp['col%']
    
# сортируем 
sorted_df = result_tmp.sort_values(
    by=['property_sys_name', 'option_sys_name', 'total universe'], 
    ascending=[True, True, False]  
)

total_row_df = pd.DataFrame([total_row], columns=result_tmp.columns)
result = pd.concat([total_row_df, sorted_df], ignore_index=True)

# округляем до нужной точности
result['col%'] = result['col%'].round(1)
for d in demo:
    result[d['demo_name']+'_col%'] = result[d['demo_name']+'_col%'].round(1)
    result[d['demo_name']+'_row%'] = result[d['demo_name']+'_row%'].round(1)
    result[d['demo_name']+'_affinity'] = result[d['demo_name']+'_affinity'].round(0)

result
```
