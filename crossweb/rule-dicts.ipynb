# Справочник по медиа-дереву Cross Web

Данный справочник разработан для определения иерархии медиаобъектов в проекте Cross Web с учетом доменов и кодов приложений. Ниже приведены примеры работы со справочником и настройки фильтров, которые помогут оптимизировать поиск нужных параметров.

### Инициализация
При построении справочника первый шаг в любом ноутбуке - загрузка библиотек, которые помогут обращаться к API Cross Web и работать с данными.

Выполните следующую ячейку, для этого перейдите в нее и нажмите Ctrl+Enter

```python
import pandas as pd

from mediascope_api.crossweb import tasks as cwt
from mediascope_api.crossweb import catalogs as cwc

pd.set_option('display.max_columns', None)
# Задаем максимальное количество выводимых строк. Раскомментируйте нужную строку
# 200 строк
# pd.set_option("display.max_rows", 200)
# Отображаем все строки. ВАЖНО! Отображение большого DataFrame требует много ресурсов
# pd.set_option("display.max_rows", None)

# Cоздаем объекты для работы с API Cross Web
mtask = cwt.CrossWebTask()
cats = cwc.CrossWebCats()
```

### Параметры, используемые для поиска по справочнику:

- holding_ids - CrossMediaHoldingID
- holding - CrossMediaHoldingName
- resource_ids - CrossMediaResourceID
- resource - CrossMediaResourceName
- product_ids - CrossMediaProductID
- product - CrossMediaProductName
- theme_ids - CrossMediaThemeID (тематики на уровне медиапродуктов)
- theme - CrossMediaThemeName (тематики на уровне медиапродуктов)
- resource_theme_ids - CrossMediaResourceThemeID (тематики на уровне ресурсов)
- resource_theme - CrossMediaResourceThemeName (тематики на уровне ресурсов)
- product_url - домены/url, по которым осущетсвляется сбор данных по тому или иному медиапродукту
- product_bundle_id - коды приложений (bundleID), по которым осущетсвляется сбор данных по тому или иному медиапродукту
- product_date_from - дата начала действия правил сбора данных по тому или иному домену/url или коду приложений
- product_date_to - дата окончания действия правил сбора данных по тому или иному домену/url или коду приложений

### 1. Поиск в справочнике медиа-дерева по заданным параметрам
Если при поиске не используются все параметры, укажите значение "None" для неиспользуемых полей.

```python
# Ввод требуемых параметров
media_rule_filter = cats.get_media_rule_tree(
    holding_ids=None,
    resource_ids=[3768],
    product_ids=[14502],
    theme_ids=None,
    resource_theme_ids=None,
    holding=None,
    resource=None,
    product=None,
    theme=None,
    resource_theme=None,
    product_url=None,
    product_bundle_id=None,
    product_date_from=None,
    product_date_to=None,
    order_dir="DESC",
    offset=None,
    limit=None,
    use_cache=True)

media_rule_filter
```

```python
# Сохранить в Excel:
writer = pd.ExcelWriter('media_rule_filter.xlsx')
media_rule_filter.to_excel(writer, 'media_rule_filter', index=False)
writer.close()
```
